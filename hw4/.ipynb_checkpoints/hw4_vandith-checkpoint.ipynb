```python
import sys
import string
import copy
from collections import Counter
from operator import itemgetter
import numpy as np
import pandas as pd
import random
import math
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
```

```python
def process_str(s):
    return s.translate(None, string.punctuation).lower().split()

def read_dataset(file_name):
    dataset = []
    with open(file_name) as f:
        for line in f:
            index, class_label, text = line.strip().split('\t')
            words = process_str(text)
            dataset.append( (int(class_label), set(words)) )
    return dataset

def get_most_commons(dataset, skip=100, total=1000):
    my_list = []
    for item in dataset:
        my_list += list(item[1])

    counter = Counter(my_list)

    temp = counter.most_common(total+skip)[skip:]
    words = [item[0] for item in temp]
    return words

def generate_vectors(dataset, common_words):
    d = {}
    for i in range(len(common_words)):
        d[common_words[i]] = i
    vectors = []
    for item in dataset:
        vector = [0] * len(common_words)
        for word in item[1]:
            if word in d:
                vector[d[word]] = 1
        vectors.append( (item[0], vector) )
    return vectors

def naive_bayes_learn(train_vectors):
    likelihoods = []
    priors = [0.0, 0.0]
    train_vector_length = len(train_vectors[0][1])
    for vector in train_vectors:
        priors[ vector[0] ] += 1
    summed = sum(priors)
    priors[0] = priors[0] / summed
    priors[1] = priors[1] / summed
    for i in range(train_vector_length):
        likelihood = [[1.0, 1.0], [1.0, 1.0]] # class, value
        for vector in train_vectors:
            likelihood[vector[0]][vector[1][i]] += 1 
        summed1 = sum(likelihood[0]) 
        summed2 = sum(likelihood[1])
        likelihood[0][0] = likelihood[0][0] / summed1 
        likelihood[0][1] = likelihood[0][1] / summed1 
        likelihood[1][0] = likelihood[1][0] / summed2 
        likelihood[1][1] = likelihood[1][1] / summed2 
        likelihoods.append(likelihood)      
    return priors, likelihoods

def naive_bayes_classify(priors, likelihoods, vector):
    posterior = copy.deepcopy(priors)
    for index in range(len(vector)):
        for i in range(2): #class 0 or 1
            posterior[i] = posterior[i] * likelihoods[index][i][vector[index]]
    return posterior.index(max(posterior))

def naive_bayes_train_test(train_vectors, test_vectors):
    mistake = 0.0
    priors, likelihoods = naive_bayes_learn(train_vectors)  
    for vector in test_vectors:
        classified = naive_bayes_classify(priors, likelihoods, vector[1])
        if classified != vector[0]:
            mistake += 1       
    return mistake / len(test_vectors)

def read_dataset_2(file): #for k-fold
    dataset = []
    for line in file:
        index, class_label, text = line.strip().split('\t')
        words = process_str(text)
        dataset.append( (int(class_label), set(words)) )
    return dataset
```

```python
def read_dataset_1(file_name):
    dataset = []
    with open(file_name) as f:
        for line in f:
            index, class_label, text = line.strip().split('\t')
            words = process_str(text)
            dataset.append( (int(class_label), words) )

    return dataset

def get_most_commons_1(dataset, skip=100, total=1000):
    counter = Counter()
    for item in dataset:
        counter = counter + Counter(set(item[1]))

    temp = counter.most_common(total+skip)[skip:]
    words = [item[0] for item in temp]
    return words

def generate_vectors_1(dataset, common_words):
    d = {}
    for i in range(len(common_words)):
        d[common_words[i]] = i

    vectors = []
    labels = []
    for item in dataset:
        vector = [0] * len(common_words)
        # Intercept term.
        vector.append(1)

        for word in item[1]:
            if word in d:
                vector[d[word]] = 1

        vectors.append(vector)
        labels.append(item[0])

    return np.array(vectors), np.array(labels)

def logistic(vector):
    return 1.0 / (1 + np.exp(-vector))

def logistic_regression(features, labels):
    n, d = features.shape
    w = np.zeros(d).T
    w_prev = np.ones(d).T
    i = 0
    # Parameters for logistic regression.
    # max_iter - Maximum iterations.
    # tol - Tolerance value.
    # alpha - Step size.
    # l - L2 regularization penalty.
    max_iter = 100; tol = 1e-6; alpha = 0.01; l = 0.01
    while True:
        if (np.linalg.norm(w - w_prev) < tol) or (i >= max_iter):
            break

        h = logistic(np.dot(features,w))
        loss_grad = (np.dot(features.T,(h - labels))) + (l * w)
        w_prev = w
        w = w - (alpha * loss_grad)
        i += 1
    return w

def logistic_pred(w, features):
    threshold = 0.5
    pred = np.where(logistic(np.dot(features,w)) >= threshold, 1, 0)
    return pred

def calc_error(pred, labels):
    error = sum(np.where(pred != labels, 1, 0))
    return (error / float(labels.size))

def svm(features, labels):
    # test sub-gradient SVM
    total = features.shape[1]
    lam = 1.; D = total
    x = features; y = (labels-0.5)*2
    w = np.zeros(D); wpr = np.ones(D)
    eta = 0.5; lam = 0.01; i = 0; MAXI = 100; tol = 1e-6
    while True:
        if np.linalg.norm(w-wpr) < tol or i > MAXI:
            break
        f = np.dot(w,x.T) 
        pL = np.where(np.multiply(y,f) < 1, np.dot(-x.T,np.diag(y)), 0)
        pL = np.mean(pL,axis=1) + lam*w
        wpr = w
        w = w - eta*pL
        i += 1
    return w

def svm_pred(w, features):
    return np.where((np.dot(features,w)) >= 0, 1, 0)
```

```python
def gini_score(vectors):
    classifiers = [vectors[x][0] for x in range(0,len(vectors))]    
    probability_1 = sum(classifiers) / float(len(classifiers))
    probability_0 = 1.0 - probability_1
    gini_score = 1.0 - ( (probability_1**2.0) + (probability_0**2.0) )
    return gini_score

def splitTree(vectors):
    #calculate the gini gain for 
    gini_gain = [0]*len(vectors[0][1])
    features = [vectors[x][1] for x in range(len(vectors))]
    starting_gini = gini_score(vectors)
    for i in range(len(gini_gain)):
        subset_0_temp = list(vectors)
        subset_1_temp = list(vectors)
        subset_1 = [subset_1_temp[x] for x in range(0,len(subset_1_temp)) if features[x][i] == 1]
        subset_0 = list([subset_0_temp[x] for x in range(0,len(subset_0_temp)) if features[x][i] == 0])
        if (len(subset_0) == 0 or len(subset_1) == 0):
            gini_gain[i] = 0
        else:
            gini_0 = gini_score(subset_0)
            gini_1 = gini_score(subset_1)
            gini_gain[i] = starting_gini - ( (float(len(subset_0)) / float(len(vectors))) * gini_0 + (float(len(subset_1)) / float(len(vectors))) * gini_1) 
    max_index = gini_gain.index(max(gini_gain))
    subset_1 = [vectors[x] for x in range(0,len(vectors)) if features[x][max_index] == 1]
    subset_0 = [vectors[x] for x in range(0,len(vectors)) if features[x][max_index] == 0]
    return [subset_0 , subset_1 , max_index ] 

def splitTree_RF(vectors):
    gini_gain = [0]*len(vectors[0][1])
    features = [vectors[x][1] for x in range(len(vectors))]
    starting_gini = gini_score(vectors)
    for i in range(len(gini_gain)):
        subset_0_temp = list(vectors)
        subset_1_temp = list(vectors)
        subset_1 = [subset_1_temp[x] for x in range(0,len(subset_1_temp)) if features[x][i] == 1]
        subset_0 = list([subset_0_temp[x] for x in range(0,len(subset_0_temp)) if features[x][i] == 0])
        if (len(subset_0) == 0 or len(subset_1) == 0):
            gini_gain[i] = 0
        else:
            gini_0 = gini_score(subset_0)
            gini_1 = gini_score(subset_1)
            gini_gain[i] = starting_gini - ( (float(len(subset_0)) / float(len(vectors))) * gini_0 + (float(len(subset_1)) / float(len(vectors))) * gini_1)
    gini_gain_sample = random.sample(gini_gain, 31)
    max_index = gini_gain.index(max(gini_gain_sample))
    subset_1 = [vectors[x] for x in range(0,len(vectors)) if features[x][max_index] == 1]
    subset_0 = [vectors[x] for x in range(0,len(vectors)) if features[x][max_index] == 0]             
    return [subset_0 , subset_1 , max_index ]     

def build_tree(Nodes ,   depth , max_depth = 10 ,max_leaves = 10 ):
    output = splitTree(Nodes )
    if (depth+1) <= max_depth:
        if ((len(output[0]) > max_leaves) & (len(output[0]) != sum([output[0][x][0] for x in range(len(output[0]))])) ):
            output[0] = build_tree(list(output[0])  , depth+1 , max_depth )
        else:
            sum_classification_1 = sum([output[0][x][0] for x in range(len(output[0]))])
            sum_classification_0 = len(output[0]) - sum_classification_1
            if sum_classification_1 >= sum_classification_0:
                output[0] = [1]
            else:
                output[0] = [0]       
        if ((len(output[1]) > max_leaves) & (len(output[1]) != sum([output[1][x][0] for x in range(len(output[1]))])) ):
            output[1] = build_tree(list(output[1]) , depth + 1 , max_depth )
        else:
            sum_classification_1 = sum([output[1][x][0] for x in range(len(output[1]))])
            sum_classification_0 = len(output[1]) - sum_classification_1
            if sum_classification_1 >= sum_classification_0:
                output[1] = [1]
            else:
                output[1] = [0]          
    else:
        sum_classification_1a = sum([output[0][x][0] for x in range(len(output[0]))])
        sum_classification_0a = len(output[0]) - sum_classification_1a
        if sum_classification_1a >= sum_classification_0a:
            output[0] = [1]
        else:
            output[0] = [0]   
        sum_classification_1b = sum([output[1][x][0] for x in range(len(output[1]))])
        sum_classification_0b = len(output[1]) - sum_classification_1b
        if sum_classification_1b >= sum_classification_0b:
            output[1] = [1]
        else:
            output[1] = [0]           
    return output

def build_tree_RF(Nodes ,   depth , max_depth = 10 ,max_leaves = 10 ):
    output = splitTree_RF(Nodes )
    if (depth+1) <= max_depth:
        if ((len(output[0]) > max_leaves) & (len(output[0]) != sum([output[0][x][0] for x in range(len(output[0]))])) ):
            output[0] = build_tree_RF(list(output[0])  , depth+1 , max_depth )
        else:
            sum_classification_1 = sum([output[0][x][0] for x in range(len(output[0]))])
            sum_classification_0 = len(output[0]) - sum_classification_1
            if sum_classification_1 >= sum_classification_0:
                output[0] = [1]
            else:
                output[0] = [0]       
        if ((len(output[1]) > max_leaves) & (len(output[1]) != sum([output[1][x][0] for x in range(len(output[1]))])) ):
            output[1] = build_tree_RF(list(output[1]) , depth + 1 , max_depth )
        else:
            sum_classification_1 = sum([output[1][x][0] for x in range(len(output[1]))])
            sum_classification_0 = len(output[1]) - sum_classification_1
            if sum_classification_1 >= sum_classification_0:
                output[1] = [1]
            else:
                output[1] = [0]          
    else:
        sum_classification_1a = sum([output[0][x][0] for x in range(len(output[0]))])
        sum_classification_0a = len(output[0]) - sum_classification_1a
        if sum_classification_1a >= sum_classification_0a:
            output[0] = [1]
        else:
            output[0] = [0]   
        sum_classification_1b = sum([output[1][x][0] for x in range(len(output[1]))])
        sum_classification_0b = len(output[1]) - sum_classification_1b
        if sum_classification_1b >= sum_classification_0b:
            output[1] = [1]
        else:
            output[1] = [0]           
    return output  
        
def classify(tree , sentance):
    root_node = len(tree)
    if root_node == 1:
        x = tree[0]
        return x
    elif root_node == 3:
        
        word_index = tree[2]
        word_value = sentance[1][word_index]
        if word_value == 1:
            return classify(list(tree[1]) , list(sentance))
        elif word_value == 0:
            return classify(list(tree[0]) , list(sentance))
        else:
            print "Check code"      
                
def classify_all(tree , vector):
    classification = np.zeros(len(vector))
    for i in range(0 , len(vector) ):
        classification[i] = classify(tree, vector[i])  
    return classification  

def dt_loss_rate(classification_dt, classification_actual):
    loss_rate = ( sum(abs(classification_dt - classification_actual)) / float(len(classification_actual)) )
    return loss_rate
                             
def bagging(train , test , samples = 50, max_depth=10):    
    bagging_classification = np.zeros((samples, len(test) ))
    for i in range(samples):
        #print "SAMPLE " + str(i)
        sampled = [random.choice(train) for x in range(len(train))]
        tree = build_tree(list(sampled) , 0 ,max_depth)
        for k in range(len(test)):
            bagging_classification[i,k] = classify(list(tree), list(test[k]))    
    bagging_classification_aggregate, z =  sp.stats.mode(bagging_classification)    
    return bagging_classification_aggregate[0]
                                                                                            
                                                                                                                                                                
def randomForest(train , test , samples = 50,max_depth=10):
    random_forest_classification = np.zeros((samples, len(test) ))
    for i in range(samples):
        #print "RF Sample " + str(i)
        tree = build_tree_RF(list(train) , 0,max_depth )
        for k in range(len(test)):
            random_forest_classification[i,k] = classify(list(tree), list(test[k]))  
    random_forest_classification_aggregate, z =  sp.stats.mode(random_forest_classification)    
    return random_forest_classification_aggregate[0]

def boosting(weights , data,max_depth=10):
    indexes = range(len(data))                                                 
    sampled = [data[np.random.choice(indexes, p = weights)] for x in range(len(data))]   
    tree = build_tree(sampled , 0,max_depth)       
    predicted_classifications = classify_all(tree,data)       
    actual_classifications =  [data[x][0] for x in range(len(data))]  
    classification_same = abs(actual_classifications - predicted_classifications) 
    for i in range(len(predicted_classifications)  ):
        if predicted_classifications[i] == 0:
            predicted_classifications[i] = -1
        if actual_classifications[i] == 0:
            actual_classifications[i] = -1
    error_k = sum(classification_same * weights)
    alpha_k = 0.5 * (  np.log(1-error_k) - np.log(error_k+0.00001) )
    new_weights1 = np.exp((alpha_k * predicted_classifications * actual_classifications)+.00001) * weights
    new_weights = new_weights1 / sum(new_weights1)
    return alpha_k , new_weights , predicted_classifications  , tree

def boosting_all( data , test ,max_depth=10, num_ens = 50 ):
    alpha_ks = [0] * num_ens        
    weights = [1.0 / len(data)] * len(data)    
    boosting_classification = np.zeros((num_ens, len(data) ))
    test_classification = np.zeros((num_ens, len(test) ) )
    for i in range(num_ens):
        #print i
        alpha_ks[i] , new_weights , boosting_classification[i] , tree = boosting(weights , data,max_depth)
        weights = list(new_weights)
        del new_weights
        test_classification[i] = classify_all(tree, test)
        del tree   
    for i in range(len(test_classification)):
        for k in range(len(test_classification[0,:])):
            if test_classification[i,k] == 0:
                test_classification[i,k] = -1
    boosting_classification_aggregate =  np.dot(np.transpose(test_classification) , alpha_ks)
    for i in range(len(boosting_classification_aggregate)):
        if boosting_classification_aggregate[i] > 0:
            boosting_classification_aggregate[i] = 1
        else:
            boosting_classification_aggregate[i] = 0
    return boosting_classification_aggregate 
```

I am comparing the accuracies of my Decision Tree and Bagging models for different fractions of training data. My Null Hypothesis is that performance is same for both these models.

I tested my hypothesis using two tailed paired t test. I compared the distirbution of accuracies for different fractions of Training data for both the models. My t test resulted in a probability of 0.024 which is less than the alpha value 0.05. So we have enough evidence to believe that the probability of our null hypothesis being true is very low.

So I reject my null hypothesis and accept my non directional alternative Hypothesis that there is a difference in the distirbution of accuracies of Decision Tree and Bagging.

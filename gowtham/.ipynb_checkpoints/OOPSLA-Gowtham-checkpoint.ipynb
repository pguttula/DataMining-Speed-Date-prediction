```python
import scipy as sp
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import sys
import os
import matplotlib.pyplot as plt
data = pd.read_csv('canvas_offline.csv')
x = np.array(data['num ops per round'])
y = np.array(data['diff size (KB)'])
y1 = np.array(data['total size (KB)'])
y2 = np.array(data['gzip diff size (KB)'])
#print x
area = np.pi*10
#plt.scatter(x, y, s=area, color = 'purple', alpha=0.5)
plt.plot(x, y, label = 'diff size (KB)',color = 'green')
plt.plot(x, y1, label = 'total size (KB)',color = 'hotpink')
plt.plot(x, y2, label = 'gzip diff size (KB)',color = 'purple')
plt.title('canvas_offline',fontsize=24,color='purple')
plt.xlabel('num of operations per round',fontsize=14,color='purple')
plt.ylabel('size in KB',fontsize=14,color='purple')
plt.legend(loc='upper left')
plt.savefig('Fig-canvas_offline.png')
plt.show()
plt.close()
```

```python
import scipy as sp
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import sys
import os
import matplotlib.pyplot as plt
from scipy import stats
data = pd.read_csv('canvas_online.csv')
#print data
x = np.array(data[' n_ops_per_round'])
y = np.array(data[' per-round comp time'])
y1 = np.array(data[' average merge time'])
area = np.pi*10
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope*x+intercept
slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(x,y1)
line1 = -10*slope1*x
plt.plot(x,y,'o', x, line,label = 'per-round comp time')
plt.plot(x,y1,'^', x, line1,label = 'average merge time')
plt.legend(loc='upper left')
plt.title('canvas_online',fontsize=24,color='purple')
plt.xlabel('num of operations per round',fontsize=14,color='purple')
plt.ylabel('Time in micro sec',fontsize=14,color='purple')
plt.savefig('Fig-canvas_online.png')
plt.show()
plt.close()
```

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
data = pd.read_csv('rbset_online.csv')
x = np.unique(np.array(data[' n_ops_per_round']))
grouped = data.groupby(' n_ops_per_round')
y = np.array(grouped[' per-round comp time'].agg(np.mean))
y1 = np.array(grouped[' average merge time'].agg(np.mean))
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope*x+intercept
slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(x,y1)
line1 = slope1*x+intercept1
plt.plot(x,y,'o', x, line,label = 'per-round comp time')
plt.plot(x,y1,'^', x, line1,label = 'average merge time')
plt.title('rbset_online',fontsize=24,color='purple')
plt.xlabel('num of operations per round',fontsize=14,color='purple')
plt.ylabel('Time in micro sec',fontsize=14,color='purple')
plt.legend(loc = 'upper-left')
plt.savefig('Fig-rbset_online.png')
plt.show()
plt.close()
```

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
data = pd.read_csv('mlist_online.csv')
x = np.unique(np.array(data[' n_ops_per_round']))
grouped = data.groupby(' n_ops_per_round')
y = np.array(grouped[' per-round comp time'].agg(np.mean))
y1 = np.array(grouped[' average merge time'].agg(np.mean))
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope*x+intercept
slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(x,y1)
line1 = 0.5 * slope1*x+intercept1
plt.plot(x,y,'o', x, line,label = 'per-round comp time')
plt.plot(x,y1,'^', x, line1,label = 'average merge time')
plt.title('mlist_online',fontsize=24,color='purple')
plt.xlabel('num ops per round',fontsize=14,color='purple')
plt.ylabel('Time in micro sec',fontsize=14,color='purple')
plt.legend(loc='upper left')
plt.savefig('Fig-mlist_online.png')
plt.show()
plt.close()
```
